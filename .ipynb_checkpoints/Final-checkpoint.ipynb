# Procesamiento digital de señales de Electromiografía

## 1. Resumen
El presente trabajo explora diversos métodos para procesar y analizar señales de electromiografía con el objetivo de caracterizar alteraciones en unidades motoras, que se asocian a disitintas patologías. 
Para ello, se utilizaron herramientas de procesamiento digital, incluyendo filtros digitales, análisis temporal y estimación espectral. Esto permitió discriminar señales de electromiografía provenientes de registros obtenidos en tres sujetos: uno sludable, uno con neuropatía y otro con miopatía.

## 2. Introducción

### 2.1 EMG 

Un electromiograma (EMG) permite registrar la actividad eléctrica producida en el músculo esquelético durante su contracción. Por lo que facilita la evaluación de la salud de los músculos y las neuronas motoras los controlan. 

La contracción muscular inicia con la llegada del potencial de acción desde neuronas motoras, que al alcanzar la unión neuromuscular, desencadena una serie de eventos eléctricos y mecánicos en la fibra muscular. Esta actividad eléctrica puede captarse y medirse mediante una electromiografía, que permite analizar la dinámica de activación neuromuscular.



*Figura 1 Representación de una unidad motora*

En estado de reposo, la membrana de tanto las células nerviosas como las musculares posee una polaridad, que se atribuye a la diferencia de concentraciones de iones entre el medio extra e intracelular. Al recibir un estímulo nervioso, la fibra muscular se despolariza y la señal se propaga a lo largo de su membrana, lo que produce una contracción. Dicha despolarización implica un movimiento de corrientes iónicas, que genera un campo eléctrico local registrable por un electrodo.

Si bien puede parecer aleatoria, la señal de EMG puede modelarse como una superposición de potenciales de acción de unidades motoras (MUAPs) que se suman y producen una respuesta.
Esto se puede describir mediante el siguiente modelo matemático:

$ x(n) = \sum_{r=0}^{N-1} h(r)*e(n-r) + w(n) $

Donde:
$x(n)$ es la señal total del EMG. 
$e(n)$ representa los impulsos neuronales, disparos de una motoneurona. 
$h(r)$ es la respuesta del músculo al impulso (MUAP). 
$w(n)$ es el ruido.
La señal de EMG presenta diversos desafíos en cuanto a su adquisición, entre los cuales se encuentran el ruido eléctrico, las interferencias electromagnéticas, artefactos de movimiento, variabilidad fisiológica de la señal. A su vez, existen factores fisiológicos que afectan la forma y amplitud, siendo estos la ubicación de los electrodos, el tipo de fibra muscular, la cantidad de unidades motoras activas, entre otros.

La adquisición de señales electromiográficas y su análisis permite estudiar el funcionamiento del sistema neuromuscular, así como también detectar posibles patologías. 

### 2.2  EMG normal

En un EMG normal, es posible diferenciar diversos niveles de activación muscular según la actividad eléctrica registrable por los electrodos. En estado de reposo, el registro no muestra actividad eléctrica significativa, pues las fibras musculares están inactivas. Al producirse una contracción débil, se da un reclutamiento de unidades motoras, la amplitud de la onda representa el número de fibras musculares en la unidad motora. A medida que se incrementa la intensidad de la contracción, aumentan tanto la frecuencia de impulsos que arrivan a la unidad motora como el número de unidades motoras que se contraen, esto produce una señal con alta interferencia y superposición, ya que muchas unidades motoras se contraen en simultáneo.

### 2.3 Patologías
Una unidad motora se compone de una neurona que inerva a fibras musculares. Por lo que pueden detectarse patologías relacionadas al mal funcionamiento de la neurona o del músculo.

En un paciente con neuropatía, el número de unidades motoras es menor al normal debido a daño en motoneuronas o sus axones.  Esta patología resulta en una disminución del número de potenciales de acción registrados, no obstante, las unidades motoras que sobreviven suelen sufrir una hipertrofia compensatoria. Esto implica que se generan MUAPs de mayor amplitud y duración en comparación a lo esperable para un paciente sano. A su vez, en este tipo de señales se suelen observar contracciones involuntarias en reposo. 

En un paciente con miopatía, es el músculo el que deteriora su capacidad contráctil. Si bien se produce un control nervioso correcto, las unidades motoras contienen menos fibras musculares funcionales. Estas condiciones se traducen en una señal con MUAPs de baja amplitud y corta duración. Es esperable ver menor energía y actividad menos densa incluso cuando el paciente contrae voluntariamente. En cuanto a la frecuencia de actividad, puede llegar a ser más alta para compensar la debilidad muscular.
 
El objetivo del presente trabajo es explorar diversos métodos y técnicas de procesamiento digital de señales fisiológicas reales. Para lograr caracterizar y comparar señales musculares de sujetos sanos y con patologías.


## 3. Metodología

### 3.1 Señales

El presente trabajo analiza señales de EMG registradas en el músculo tibial anterior de tres sujetos con diferentes condiciones neuromusculares. Los datos se recolectaron a través de un electrodo concéntrico con aguja de 25 mm insertado en el músculo. A cada uno de los pacientes se le pidió realizar una dorsiflexión suave del pie contra una resistencia, posicionando el electrodo hasta identificar potenciales de unidades motoras con tiempo de ascenso rápido. Luego, se procedió el muestreo durante varios segundos de señal, antes de retirar la aguja, los sujetos relajaron el músculo.
Las señales fueron inicialmente adquiridas a 50kHz y remuestreadas a 4kHz. Se aplicaron además dos filtros pasa altos de 20 Hz y un pasa bajos de 5kHz. Las tres señales adquiridas corresponden a:
1. Un hombre de 44 años sin antecedentes de enfermedad neuromuscular.
2. Hombre de 62 años con dolor lumbar crónico y neuropatía por radiculopatía L5 derecha.
3. Hombre de 57 años con miopatía secundaria a polimiositis crónica, medicada con esteroides y bajas dosis de metotrexato.

### 3.2 Preprocesamiento

#### 3.2.1 Filtrado

##### Filtrado Lineal

Si bien las señales ya pasaron por etapas de prefiltrado analógico, es conveniente aplicar filtros digitales para refinar el acondicionamiento de las señales y limitar los rangos de frecuencias a un rango fisiológicamente relevante.

La actividad eléctrica muscular se encuentra mayormente concentrada entre 20 Hz y 500 Hz. Las frecuencias más bajas pueden deberse a movimientos, mientras que las superiores suelen representar interferencia o ruido eléctrico. El acondicionamiento adicional por medio de la implementación de filtros digitales resulta beneficioso, pues mejora la relación señal a ruido y permite refinar las bandas de paso y rechazo con mayor precisión.

##### Plantilla de diseño 

##### - Filtros Pasa Bandas
Para filtrar las señales de EMG, se diseñaron dos filtros de tipo IIR recursivos pasa banda. Ambos poseen una respuesta suave, que atenúa las componentes fuera del rango de 20 a 480 Hz. Se estableció una banda de rechazo por debajo de 10 Hz para suprimir artefatos debido al movimiento y una banda de rechazo en 600 Hz para eliminar componentes no fisiológicas. La ondulación máxima permitida en la banda pasante fue de 0.5 dB, y la atenuación mínima de 40dB en las bandas de rechazo. Los parámetros aseguran una buena fidelidad sin introducir distorsiones en la información relevante de la señal.



*Figura 5: Filtro pasa banda - plantilla de diseño*

Se probaron dos diseños posibles, un filtro de tipo Butterwoth y un Chebyshev tipo I. Si bien el segundo cumple con los requerimientos con un orden menor, introduce ripple en la banda pasante, por lo que se concluyó que el filtro de Butterworth (al costo de un mayor esfuerzo computacional) es una mejor opción para filtrar las señales ya que su respuesta monótona en la banda de paso no introduce ondulaciones en la señal en el rango frecuencial de interés. 

##### - Filtros Notch
Además, para eliminar la interferencia eléctrica, se diseñó un filtro notch con corte en 60 Hz. Se probó la implementación de múltiples etapas en cascada para alcanzar órdenes superiores. 



*Figura 6: Filtro notch - plantilla de diseño*

La plantilla permite ver que un orden 2  logra cumplir con los requerimientos de atenuación y transiciona en forma más abrupta.

##### - Filtros Concatenados

A continuación, se concatenaron las secciones SOS del filtro pasabanda Butterworth y el filtro notch de orden 2.



*Figura 7: Concatenación de filtros (notch + pasabanda) - plantilla de diseño*

Como se logra apreciar en la figura, el filtro diseñado atenúa las regiones que están por fuera del rango de frecuencias de interés a -40dB. Asimismo, la elección del pasabandas butter logra una transferencia sin rizado en la banda pasante. 


##### Aplicación del Filtro Diseñado

Para implementar el filtro, se utilizó la función sosfiltfilt, que aplica el filtro diseñado dos veces. Primero hacia adelante, y luego hacia atrás. La técnica de filtrado bidireccional compensa la distorsión de fase introducida. Como resultado, el retardo es nulo y la señal no se distorsiona. No obstante, no es causal, por lo que no puede utilizarse en aplicaciones en tiempo real. 
Tras diseñar e implementar el filtro digital, se graficó la señal original y filtrada en intervalos de mayor y menor ruido, observándose que la señal filtrada conserva la forma de onda original relativamente bien, pero queda removida su línea de base, a su vez, se ve una reducción en la amplitud de los picos. Por otro lado, se introudce un leve rizado en áreas específicas.

##### Filtrado no Lineal
Asimismo, se decibió explorar métodos de filtrado no lineal para comparar los resultados con los del método lineal. 
Si se considera a la señal adquirida cruda como una suma entre ruido y la señal limpia, puede construirse un filtro basado en la estimación de la señal de ruido. 
Para estimar la señal de ruido, se comenzó por utilizar la función find_peaks SciPy sobre la señal en crudo. 



*Figura 14: Detección de picos*

Por inspección visual, se determinó que aproximadamente 3 ms antes de cada pico, la señal debería tener un nivel isoeléctrico nulo. Teniendo esto en consideración, se decidió implementar la técnica de splines para estimar la línea de base interpolando los valores de la señal 3 ms antes de los picos detectados. 



*Figura 15: Detección de picos y puntos en nivel isoeléctrico nulo*

Para mejorar la técnica, en lugar de estimar la línea de base interpolando directamente los puntos hallados, se tomaron ventanas de 4 ms centradas en los puntos. Luego, se obtuvo la media y la mediana de cada ventana, a partir de estos valores se construyeron las líneas de base por interpolación de la media y mediana de cada ventana.



*Figura 16: Estimación de línea de base por interpolación de media y mediana en nivel isoeléctrico nulo*

A su vez, se incorporaron más valores a la interpolación, cuyo valor se calculó como la mediana en un intervalo proporcional al largo de la ventana. Se prefirió el método de tomar la mediana por sobre el promedio porque tiene mayor robustez ante oscilaciones por ruido de alta frecuencia.



*Figura 17: Filtrado por remoción de línea de base por interpolación mediana en nivel isoeléctrico nulo*

Por último, se restó la línea de base estimada a la señal original, esto resultó en una señal de media casi nula pero que conserva perfectamente su forma y la amplitud de sus picos.

A su vez, se probó un filtrado de mediana, este tipo de filtro tiene como fin la supresión de valores atípicos sin introducir distorsión en los bordes. Para su implementación, se definió una ventana de 5 ms, adecuada para eliminar ruido de baja frecuencia sin afectar eventos fisiológicos de larga duración.



*Figura 18: Comparativa de filtros no lineales (filtro de mediana contra filtro por interpolación)*


*Figura 19: Comparativa de filtros no lineales (filtro de mediana contra filtro por interpolación)*

Como se logra apreciar, tanto el filtro de mediana como el filtro por interpolación logran remover la línea de base sin introducir rizado o cambiar la forma de la señal original. Asimismo, la amplitud de los picos se conserva.

##### Comparación de filtros

Tras descomponer la señal, se utilizó waverec para reconstruir la señal pero esta vez dejando afuera los niveles de frecuencia no deseados.

La visualización temporal de la señal original y la señal luego de atravesar el filtro wavelet muestra resultados que combinan buena remoción de línea de base (frecuencias bajas) y preservación de la forma de onda y amplitud de picos. Cabe destacar que se probaron diversas onditas madre, para seleccionar 'dmey', que resulta en baja distorsión y facilita el posterior análisis cualitativo.

Se incorporó la señal filtrada mediante el método de onditas a la estimación de PSD. Esto permite visualizar que, si bien no produce una transición tan abrupta y atenuación tan buena como el filtro de Butterworth, la atenuación es aceptable (aproximadamente de 40dB).

##### Filtro Final

El diseño e implementación de los filtros permitió comparar diversas opciones de acondicionamiento para señales EMG. 
Tras evaluar el impacto de los filtros tanto en tiempo como en frecuencia, se decidió que, dentro de los filtros lineales, el de Butterworth ofreció una buena atenuación por fuera del rango espectral de interés, manteniendo la morfología de la señal. Además, su gran simplicidad y eficiencia computacional lo hace una excelente opción para procesar grandes bases de datos. En particular, su implementación en secciones SOS asegura estabilidad numérica y bajo costo por muestra. En cuanto al filtrado bidireccional, asegura la preservación de la morfología original pero implica una operación no causal.
En cuanto a los filtros no lineales, incluyendo el filtro por interpolación y el de mediana, demostraron ser útiles para remoción de línea de base, manteniendo la morfología de la señal. Sin embargo, no ofrecieron buena atenuación de frecuencias altas. Por otro lado, el filtro diseñado por el **método de onditas** combinó buena atenuación de frecuencias altas y bajas, con baja distorsión en tiempo. 
Se concluyó entonces, que las opciones más adecuadas fueron el filtro de Butterworth y el de onditas (estos dos fueron implementados a lo largo de las secciones posteriores según la convenciencia). No obstante, la elección del filtro depende de los fines particulares de cada estudio.


#### 3.2.2 Rectificación de señales y envolventes

  
Al procesar una señal de EMG, se consideran (comúnmente) sus valores positivos pues estos reflejan la magnitud de la actividad independientemente de la dirección del potencial eléctrico. En este contexto, se puede realizar una rectificación de onda completa o de media onda. En el caso de la onda completa, se toma el valor absoluto de cada punto, además, se resta el valor medio para eliminar posibles componentes de continua que pueden distorsionar análisis posteriores.
Además, en el procesamiento de señales de EMG es habitual obtener la envolvente, ya que facilita el análisis de la actividad muscular. La envolvente produce una curva suave que sigue las variaciones de amplitud de la señal original, esto permite comparar la actividad muscular global de manera sencilla. 

Para este caso, se estimó la envolvente de las tres señales filtradas por medio de tres métodos.

##### Promediador sobre señal rectificada (ARV móvil)

El primer método consiste en estimar la envolvente aplicando una media móvil en forma directa sobre la señal rectificada, resultando en una estimación de la amplitud media local. Si bien es una técnica simple y de bajo costo computacional, no es la más popular en el análisis de señales de EMG pues no capta con alta sensibilidad los picos de alta intensidad pues no trabaja sobre la energía de la señal.
El procedimiento consiste en:
1. Rectificar la señal:
3. Aplicar un filtro promediador (media móvil).




*Figura 23: Rectificación y envolvente mediante ARV*

Matemáticamentem la etapa de promediado constituye una convolución entre la señal y un kernel rectangular de longitud M. La elección M incide directamente en la resolución temporal y el suavizado de la señal. Una ventana muy corta produce una mayor resolución en tiempo, mientras que una ventana mayor resulta en una envolvente más suave y estable. Para este caso, se utilizó una ventana de 50 ms, que ofrece un buen balance entre suavizado y resolución.
   
##### Envolvente LPF
El segundo método estima la envolvente por medio de la implementación de un filtro pasa bajos digital sobre la señal previamente rectificada. Esta técnica es de común uso en análisis de EMG pues permite suavizar en forma eficaz la señal sin alterar la morfología original. A diferencia del método de ARV móvil, en esta técnica puede regularse con precisión la frecuencia de corte, lo que permite decidir el nivel de suavidad de la envolvente. 
1. Rectificar la señal (restar la media y tomar valor absoluto).
2. Aplicar un filtro pasabajos Butterworth.



*Figura 24: Rectificación y envolvente mediante filtro pasa bajos*
Como se logra apreciar, la varación de la frecuencia de corte permite obtener una envolvente con mayor resolución (fc = 40Hz) o más suave (fc = 5 Hz).

##### Envolvente por RMS móvil

Este método produce una estimación de la magnitud de la energía local de la señal, si bien tiene un costo computacional mayor, su sensibilidad a picos intensos es mayor. El procedimiento consiste en:
1. Obtener el cuadrado de la señal rectificada. 
2. Promediar los valores en una ventana móvil de un cierto ancho.
3. Tomar la raíz cuadrada del resultado.
Este procedimiento puede interpretarse como la aplicación de un filtro FIR deslizante de media móvil sobre la señal elevada al cuadrado. A diferencia de aplicar directamente un filtro prmediador, este procedimiento refleja la magnitud muscular a lo largo del tiempo en forma más adecuada en lugar de simplemente suavizar su forma.



*Figura 25: Rectificación y envolvente mediante ventana RMS*

Nuevamente, la varaición del largo de la ventana influye en la suavidad de la envolvente, para este caso, se eligió una ventana de 80 ms, que ofrece buen compromiso entre resolución y suavidad.

Los tres métodos implementados son ampliamente utilizados en procesamiento de señales de EMG. La media móvil es una técnica simple y de bajo costo computacional, es adecuada para monitoreos rápidos o sistemas con recursos limitados. En cuanto al filtro pasa bajos, la envolvente que se produce es suave y sigue la morfología original, en consecuencia, es altamente recomendable para análisis de sincronización y activación. Por último, el método de RMS móvil proporciona una medida energética de señal lo cual resulta útil para discriminar densidad de actividad muscular. A lo largo del trabajo se utilizarán el método de RMS ya que es el más popularmente empleado para medir actividad muscular.

### 3.3 Caracterización temporal
Tras realizar un preprocesamiento completo de las señales de EMG, incluyendo filtrado (filtro wavelet), rectificación (onda completa) y estimación de la envolvente (método RMS), se procedió con un análisis comparativo temporal de las señales bajo estudio.

#### 3.3.1 Normalización 
Para realizar comparaciones válidas entre diferentes sujetos y condiciones, es habitual realizar una normalización de la señal. Esto permite compensar diferencias (como masa muscular, ubicación de electrodos y nivel de esfuerzo), para lograr interpretación relativa de la activación muscular. 

Para normalizar las señales, es común representar la señal total como porcentajes relativos a la máxima contracción voluntaria (MVC) documentada para cada sujeto en ese grupo muscular. Este tipo de normalización es ampliamente utilizada pues la señal de EMG varía entre personas, entre músculos e incluso entre sesiones, ya que depende de la colocación de electrodos, piel, fatiga, etc. La normalización permite comparar señales entre distintos sujetos o distintas condiciones y evita comparaciones injustas. Sin embargo, para llevar a cabo esta técnica de normalización, es necesario disponer de información sobre la MVC de cada sujeto en condiciones controladas. 

Ante la falta de documentación sobre la MVC, se compararon las envolventes de las señales rectificadas y filtradas, sin normalizar.

#### 3.3.3 Anlálisis cuantitativo temporal

##### Métricas generales

Para obtener los datos cuantitativos de las señales bajo estudio, se realizó una división en 3 regiones de interés. La primer región corresponde al intervalo de tiempo durante el cual el paciente realiza fuerza, el segundo intervalo es un subsegmento dentro del primero que comprende la actividad más intensa, el periodo final comprende la relajación muscular. Para las señales del paciente saludable y del paciente con neuropatía, se establecieron los segmentos por inspección visual. En cuanto a la señal de miopatía, no fue posible realizar el análisis directamente debido a la morfología de la señal, cuya amplitud es tan baja que no permite discriminar entre periodo de contracción y relajación. 

Para lograr el análisis de la señal de miopatía, se implementó el método de densidad de *turns*, que calcula cambios de dirección de la señal usando una ventana deslizante. Para aplicar el método, se calcularon las diferencias entre muestras consecutivas y se obtuvieron sus signos para así detectar cada cambio de signo como un giro o turn. Luego se compararon los cambios de signo con un umbral mínimo establecido para discriminar turns inválidos. Esto se realizó sobre el largo total de la señal en ventanas de duración fija. Mediante este método, se identificó el periodo de contracción en base a la densidad de giros, pues durante la contracción se incrementa el número de unidades motoras activas y su frecuencia de disparo, lo cual genera un mayor número de giros.

Habiendo identificado las zonas de interés en la señal de miopatía, se calcularon métricas sobre las regiones de contracción leve, contracción fuerte y relajación.

Asimismo, se observó que la amplitud media del periodo de relajación tuvo una mayor amplitud que el periodo de contracción moderada, esto se debe a la actividad espontánea presente durante la relajación. 
Para la señal obtenida del paciente con miopatía, se obtuvieron amplitudes bajas y duraciones cortas, lo cual es esperable para esta condición. 

##### NUBE Turn-Amplitude

Se continuó explorando el método de turns debido a que es muy comúnmente utilizado en análisis temporal de EMG. Este método permite analizar las señales sin necesidad de detectar potenciales individuales ni hacer descomposición. Para analizar los turns en una señal, se divide la señal en segmentos cortos, sobre los cuales se cuenta el número de turns, la amplitud entre pares de turns consecutivos y su amplitud promedio. Esta ténica produce un punto por segmento, donde sus coordenadas representan el número de turns y su amplitud promedio, resultando en una nube TA (Turn-Amplitude). Esta técnica permite describir la complejidad y fuerza de la señal sin necesidad de descomponer la señal en sus MUAPs, haciéndola una opción robusta, rápida y objetiva para analizar EMG.

En el eje horizontal, se observa la cantidad de turns por cada segmento (250 ms). Un mayor número de turns sugiere una señal más compleja, es decir, mayor reclutamiento de unidades motoras. En cuanto al eje vertical, este revela la amplitud promedio entre turns consecutivas, a mayor valor, mayor voltaje, esto sugiere contracciones más fuertes o potenciales más grandes.



*Figura 31: Nube TA* 

Como era de esperarse, se observaron valores concentrados en la zona de abajo a la derecha para la señal con miopatía, lo que sugiere muchas fibras pequeñas activas, un alto número de turns pero de baja amplitud. Para el caso de la señal del sujeto con neuropatía, se obtuvo una nube concentrada en la parte superior izquierda, con amplitudes altas pero menor número de turns, esto es esperable ya que hay pocas unidades motoras pero de mayor tamaño.



*Figura 32: Nube TA sobre segmentos de interés* 

Para enriquecer el análisis, se extrajeron segmentos de interés y se repitió el procedimiento sobre estos. Para el paciente con neuropatía, se observó una amplitud mayor, que alcanzó valores muy elevados en el periodo de contracción fuerte y final. El número de turns fue inicialmente bajo, creció e el periodo de contracción fuerte y disminuyó nuevamente en el periodo final. Esto sugiere un aumento del reclutamiento durante la contracción. Cabe destacar que los valores elevados de turns y amplitud en el segmento de relajación se atribuyen a la contracción involuntaria que suele darse en pacientes con esta patología. 
Para el paciente sano, el número de turns se mantuvo constante. En cuanto a la amplitud, creció ligeramente durante la contracción fuerte y luego bajó. Por último, el paciente con miopatía mantuvo una aplitud baja, aunque ligeramente mayor en los periodos de contracción débil y fuerte. En cuanto al número de turns, este incrementó drásticamente durante la relajación.

### 3.4 Análisis espectral
Se exploraron diversas técnicas para identificar las diferencias en el espectro de las tres señales bajo análisis.
#### 3.4.1 Análisis Cualitativo Espectral

##### Periodograma con ventana Hamming
El método de periodograma permite estimar la densidad espectral de potencia a través de la transformada de Fourier. Consiste en obtener la transformada de Fourier de una señal y a partir de ella estimar la PSD tomando por fundamento el teorema de Parseval. El truncamiento de la secuencia temporal equivale a realizar un producto con una ventana temporal de valor unitario constante, esto introduce ambigüedades en su representación en el dominio espectral. Dicha ambigüedad puede ser mitigada incrmentando la longitud de la ventana o eligiendo una ventana de forma apropiada. 
Para implementar este método, se utilizó la ventana Hamming, que ofrece un buen balance entre ancho de lóbulo principal y atenuación de lóbulos secundarios. Lo que implica buena resolución y poca varianza.



Como se logra apreciar, el periodograma permite observar la distribución de la energía en frecuencia. No obstante, se observa una alta varianza que no permite discriminar claramente entre los espectros de las señales bajo estudio.


##### PSD por Multitaper (Thompson)
Para mitigar la varianza y mejorar la visualización de la PSD, se implementó otro método de estimación de PSD, el método Multitaper. Esta técnica promedia múltiples periodogramas calculados multiplicando la señal por K tapers DPSS (ventanas ortogonales) que se diseñan para lograr una mayor concentración de energía. En comparación a la estimación de PSD con una única ventana, esta técnica reduce la varianza y el leakage.



Se graficaron ambos métodos de estimación espectral en una única imagen. Esto permitió comparar la mejoría en la varianza al estimar PSD mediante la técnica de multitaper. Si bien se mantiene la forma general del espectro, el segundo método implementado permite una visualización más clara.


##### Análisis por segmentos de interés

Este comportamiento es esperable y refleja potenciales de unidad motora normales con buena velocidad de conducción.

Si bien en potenciales individuales la frecuencia es alta debido a su corta duración, lo cierto es que en la señal compuesta, (que es resultado de la superposición de múltiples unidades motoras de baja amplitud y poco sincronizadas), se observa un espectro concentrado en bajas frecuencias, reflejando un promedio temporal lento. 
En el paciente con neuropatía se observa un claro pico dominante alrededor de 150-200 Hz y una segunda elevación en frecuencias más bajas, este patrón, denominado bimodal es típico en sincronización anómala de grandes unidades motoras, que generan potenciales prolongados y de alta amplitud.

Para el paciente con miopatía, se observa una forma de onda muy similar, sin aumento notorio en la potencia. Es decir que el esfuerzo del paciente no logra ampliar la potencia total ni el contenido de alta frecuencia, lo que coincide con la condición del paciente.
En canto al paciente con neuropatía, se ve un aumento en la potencia global, con predominio de energía en las frecuencias bajas. En comparación el periodograma con ventana, el método Multitaper ofrece una estimación de PSD más fácil de intepretar debido a la notable mejoría en la varianza de las curvas, lo que permite analizar los espectros en los intervalos de interés y discriminar las características presentes en diferentes patologías. ##### Métricas sobre PSD
Para respaldar el análisis cualitativo, se obtuvieron métricas de interés en el espectro. A partir de las curvas de densidad espectral de potencia estimadas mediante el método Multitaper.
En primer lugar, se obtuvo la energía total, la energía del espectro a través de una integración numérica utilizando la función np.trapezoid, que permite estimar la potencia total.
A continuación, se determinó la frecuencia de percentil 95 %, que revela el rango en donde se concentra el 95 % de la energía total.
A su vez, se obtuvieron la frecuencia media, que representa un promedio ponderado de frecuencia por potencia y la frecuencia mediana, que divide la potencia acumulada en dos mitades iguales.
Nuevamente, el estudio se realizó por intervalos (actividad leve, intensa y reposo) ya que pemite comparar según el nivel de activación muscular.



El primer intervalo de interés corresponde al periodo de actividad leve. 

Como se logra apreciar, el paciente sano presenta una frecuencia media relativamente baja de 130 Hz y una mediana de 91 Hz. Estos valores son relativamente bajos, lo cual es esperable ya que los potenciales de unidad motora tienen duración corta y velocidad de conducción alta, con energía que se concentra entre 100 y 400 Hz. El ancho de banda estimado mediante el percentil 95%, que fue de 344 Hz, lo cual es consistente con lo anterior. En cuanto a la energía total, el valor obtenido fue el más bajo en comparación a los otros sujetos, siendo este 0,0032 mV², lo que refleja actividad muscular global baja, acorde con un periodo en el cual se contrae débilmente el músculo.
En cuanto cuanto al paciente con miopatía, tanto la media como la mediana fueron mayores que en el paciente sano, con valores de 162 Hz y 128 Hz respectivamente. En cuanto al ancho de banda, el espectro se concentra en frecuencias más altas (freq95 = 378 Hz), y la energía total fue de 0,0055mV². Esto refleja una contracción más densa (mayores frecuencias) y de baja potencia.
En cuanto al paciente con neuropatía, los valores de media y mediana se encontraron entre los valores del paciente sano y con miopatía, siendo estos 195 Hz y 186 Hz respectivamente. La energía fue notoriamente mayor, con un valor de 0,027mV² y el ancho de banda fue el mayor, con 377 Hz. Los valores sugieren presencia de MUAPs de gran amplitud y actividad alta incluso en reposo, pudiendose atribuir a reinervación o descargas espontáneas.



En el periodo de contracción intensa, el sujeto sano presentó una media de 109 Hz y una mediana de 84 Hz, con un ancho de banda de 268 Hz. En cuanto a la energía total, esta aumentó notablemente respecto al periodo de actividad leve, alcanzandose un valor de 0,013mV², es decir, cuatro veces mayor respecto al valor en contracción muscular débil. Las métricas sugieren mayor activación muscular y reclutameinto de más unidades motoras.
El paciente con miopatía presentó una frecuencia media y mediana similar a las de reposo, con valores de 162 Hz y 126 Hz. En cuanto al ancho de banda, se estimó un valor de 380 Hz, nuevamente, similar al valor de actividad leve. Por último, la energía total estimada fue ligeramente mayor, con un valor de 0,0064 mV². Esto coincide con la fisiología miopática, en donde el músculo incrementa muy poco su potencia incluso en esfuerzos mayores, debido a pérdida de fibras musculares y la contracción poco sincronizada. 
Por último, el paciente con neuropatía presentó una frecuencia media de 140 Hz y una mediana de 112 Hz, ambos valores fueron inferiores respecto a los hallados en actividad leve, lo que corresponde con la patología, pues tiende a presentar menos MUAPs pero de mayor tamaño, por lo que las frecuencias tienden a valores más bajos. Del mismo modo, el ancho de banda fue de 331 Hz, es decir que se concentró la energía en valores de frecuencia más bajos. ´Por último, la energía total fue de 0,35 mV², su alto valor se atribuye a potenciales de gran amplitud y reclutamiento poco regular.



En reposo, las frecuencias del sujeto sin patologías tendieron a valores mayores respecto a los periodos de contracción (leve e intensa), con una media de 141 Hz y una mediana de 102 Hz. En cuanto al ancho de banda, el valor estimado fue de 352 Hz. Por último, la energía total fue la mínima, con 0,003 mV². Los valores coinciden con lo esperable para un periodo de reposo, en donde la actividad es baja.
Para el paciente con miopatía, los valores se mantuvieron relativamente constantes, con 164 Hz de media, 135 Hz de mediana, y un ancho de banda de 373 Hz. En cuanto a la energía total, se estimó un valor de 0,0056 mV². Los valores indican que no hubo actividad espontánea, y que el músculo permaneció en reposo, manteninedo su característico espectro de altas frecuencias.
El paciente con neuropatía tuvo un comportamiento anómalo. Exhibiendo un valor de freucuencia media de 182 Hz, y una mediana de 173 Hz. El ancho de banda estimado fue de 366 Hz y la energía total fue de 0,121 Hz. Los valores sugieren actividad eléctrica durante el reposo, no obstante, esto es común en neuropatías y puede atribuirse a contracción espontánea, involuntaria. 

Las métricas reflejan comportamientos esperables, característicos según el sujeto y la actividad. 
El paciente saludable mostro comportamiento fisiológico normal, energía proporcional al esfuerzo y reposo sin evidencia de actividad eléctrica.
El paciente con miopatía mostró frecuencias altas y energía muy baja a lo largo de los tres periodos, lo que encaja con lo esperable: poca potencia total debido a la reducción en el número de fibras activas, pero contenido espectral rápido que se atribuye a unidades motoras pequeñas y de breve duración.
En cuanto al comportamiento del paciente con neuropatía, refleja un músculo con reclutamiento muy irregular pues se observó actividad incluso en el reposo, lo que sugiere contracciones involuntarias (comunes en neuropatía). Esto explica el desplazamiento hacia frecuencias altas.

Las diferencias observadas pueden atribuirse a la naturaleza no estandarizada del esfuerzo durante la adquisición de las señales.


## 4. Conclusiones

El trabajo realizado permitió explorar diversas técnicas de análisis y discriminación de señales de electromiograma. Se implementó un pipeline completo y adaptativo para el análisis de EMG que abarca: preprocesamiento, análisis temporal y análisis espectral.
En la primer etapa, se diseñaron se implementaron diversos filtros digitales sobre las señales bajo estudio. Permitiendo restringir las señales al ancho de bada fisiológico esperable para electromiograma. Se probaron filtros lineales y no lineales para atenuar frecuencias inferiores a 20 Hz y superiores a 450 Hz. Se analizaron los filtros sobre las señales, para contrastar la conservación de la morfología de la señal, la atenuación de frecuencias fuera de rango, el rizado, y el costo computacional y facililidad de implementación. Finalmente, se determinó que para este estudio en particular, el filtro de Butterworth y el de Onditas fueron superiores gracias a su facilidad de uso y adaptabilidad, pero con buena atenuación y conservación de la forma de onda original de las señales.
Continuando con el preprocesamiento, se realizó una rectificación de onda completa y se probaron diversas técnicas para estimar la envolvente de las señales. Se decidió que la envolvente por RMS captura mejor la energía local y es de común uso en análisis de EMG.
A partir de las señales preprocesadas, se procedió estudiar las características temporales de cada señal, realizando una inspección inicial visual, seguido del método de nube TA para segmentar intervalos de actividad (leve, intensa y reposo). Sobre los distintos intervalos se analizaron métricas para lograr una comparación cuantitativa entre sujetos. Cabe destacar que debido a la carencia de estudio de MCV de os pacientes, no se normalizaron las señales, lo que implica una dificultad al comparar diferentes sujetos.
Por último, se analizó el espectro de las señales. Se compararon el método de periodograma con ventana Hamming con el de Multitaper, permitiendo concluir que el segundo lograba estimar la densidad espectral de potencia en forma más nitida gracias a la baja varianza. Asimismo, se obtuvieron métricas sobre la PSD estimada para analizar el contenido espectral en los intervalos de contracción leve, intensa y reposo de cada sujeto.
Las etapas realizadas permiten un análisis profundo y robusto de señales de electromiografía, que permiten analizar y discriminar el comportamiento de pacientes con diversas patologías.
